```python
# Load each quiz file
quiz1 = pd.read_csv('./data/quiz_1_grades.csv', index_col='Email')
quiz2 = pd.read_csv('./data/quiz_2_grades.csv', index_col='Email')
quiz3 = pd.read_csv('./data/quiz_3_grades.csv', index_col='Email')
quiz4 = pd.read_csv('./data/quiz_4_grades.csv', index_col='Email')
quiz5 = pd.read_csv('./data/quiz_5_grades.csv', index_col='Email')
```

```python
quiz1.head()
```

```python
# List of quizzes
quizzes = [quiz1, quiz2, quiz3, quiz4, quiz5]
quiz_names = ['Quiz 1', 'Quiz 2', 'Quiz 3', 'Quiz 4', 'Quiz 5']

for i, quiz in enumerate(quizzes):
    # Drop 'First Name' and 'Last Name' columns
    quiz = quiz.drop(columns=['First Name', 'Last Name'], errors='ignore')
    # Rename remaining columns with the quiz name
    quiz.columns = [f"{col}_{quiz_names[i]}" for col in quiz.columns]
    
    # Remove 'Grade_' prefix from column names
    quiz.columns = [col.replace('Grade_', '') for col in quiz.columns]
    
    quizzes[i] = quiz  # Update the list with the renamed DataFrame

# Merge all quizzes into a single DataFrame
quiz_data = quizzes[0]
for quiz in quizzes[1:]:
    quiz_data = quiz_data.join(quiz, how='outer')

# Display the combined quiz data
quiz_data.head()
```
